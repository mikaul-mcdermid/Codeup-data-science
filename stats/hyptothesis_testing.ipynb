```python
df2 = data('mpg')
df2.head()
alpha = 0.05
```

```python
# H0 = no relation between manufacturer and cylinder type
# Ha = there is a relation between manufacturer and cylinder type
```

```python
df2.manufacturer.value_counts()
df2.cyl.value_counts()
observed2 = pd.crosstab(df2.manufacturer, df2.cyl)
chi2, p, dof, expected = stats.chi2_contingency(observed2)
print('Observed')
print(observed2)
print('\nExpected')
print(expected.astype(int))
print('\n----')
print(f'chi^2 = {chi2:.4f}')
print(f'p     = {p:.4f}')
```

```python
if p < alpha:
    print('reject the null hypothesis')
else:
    print('fail to reject the null hypothesis')
```
